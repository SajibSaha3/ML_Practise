```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
# load the dataset 
iris = datasets.load_iris()
iris
```

```python
x = iris.data
y = iris.target
target = iris.target_names
```

```python
target
```

```python
# Feature scaling through standardscaler
st = StandardScaler()
x_scale = st.fit_transform(x)
# Assuming x_scale is your NumPy array
for row in x_scale:
    print(' '.join(f'{value:.2f}' for value in row))
```

```python
y
```

```python
# Apply PCA
pca = PCA(n_components=2)
new_x = pca.fit_transform(x_scale)
new_x
```

```python
# plot pca
plt.figure()
colors =["red", "blue", "yellow"]

for color, i, target_name in zip(colors, [0,1,2], target):
    plt.scatter(new_x[y == i, 0], new_x[y == i, 1], color = color, alpha = .8, lw = 2, label= target_name)

plt.legend(loc = "best", shadow = False, scatterpoints = 1)
plt.title("PCA for irsi data")
plt.show()
```

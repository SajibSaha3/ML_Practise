```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Linear dataset

```python
df = pd.read_csv("linear_data.csv") # load the linear datset
```

```python
df.head() # see the 1st 5 data
```

```python
#seperate two columns into two variable
linear_x = df[["x"]] 
linear_y = df[["y"]]
```

# non linear dataset

```python
df1 = pd.read_csv("nonlinear_data.csv")# load the non linear dataset 
```

```python
df1.head() # see the 1st 5 data
```

```python
#seperate two columns into two variable
non_linear_x = df1[["x"]]
non_linear_y = df1[["y"]]
```

# visual representation of data sets

```python
# scatter plot for linear dataset
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.scatter(linear_x, linear_y, color = "red", label ="linear_relationship" )
plt.xlabel("X_axis")
plt.ylabel("Y_axis")
plt.title("Linearly related")
plt.legend()

# scatter plot for non-linear dataset
plt.subplot(1, 2, 2)
plt.scatter(non_linear_x, non_linear_y, color='black', label='non_linear_relationship')
plt.xlabel('X_axis')
plt.ylabel('Y_axis')
plt.title('Non-Linearly related')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# see the internal correlation between for linear data 
df.corr()
```

```python
# see the internal correlation between  for non linear data
df1.corr()
```

# perform Linear Regression

```python
# import linear regression
from sklearn.linear_model import LinearRegression
lr = LinearRegression() # create object for model
```

```python
# train the model with linear data 
lr.fit(linear_x,linear_y)
```

```python
# test the performance score 
lr.score(linear_x,linear_y)
```

```python
predicted = lr.predict(linear_x)
```

# visualize the best fit line

```python
plt.scatter(linear_x,linear_y, color="black",  label= "Linear dataset Relation")
plt.plot(linear_x, predicted, color="red", label= "Best fit line")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Best fit line with linear data")
plt.legend()
```

# Linear regression with non linear data

```python
# train the model with data  
lr.fit(non_linear_x, non_linear_y) 
```

```python
# test the accuracy
lr.score(non_linear_x, non_linear_y)
```

```python
# predict value for y_pred
predicted = lr.predict(non_linear_x)
```

# visualize best fit line for non linear data with Linear Regreesion

```python
plt.scatter(non_linear_x, non_linear_y, color="Green",  label= "non Linear dataset Relation")
plt.plot(non_linear_x, predicted, color="red", label= "Best fit line")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Best fit line with non linear data")
plt.legend()
```

# polynominal Regression for non linear data

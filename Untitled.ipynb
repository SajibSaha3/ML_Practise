```python
# Spilitting the data set into train test 
from sklearn.model_selection import train_test_split
```

# Linear model

```python
#Linear model regression
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(xtrain,ytrain) # training the model
```

```python
reg.predict([[70.865815]])
```

```python
m = reg.coef_ 
```

# Best fit line

```python
plt.scatter(df[['Height']], df[['Weight']] , color='green', marker='+')
plt.plot(df.Height , reg.predict(df[['Height']]),color='red') # best fit line
plt.xlabel("Height in inch")
plt.ylabel('Weight')
plt.title('Weight based on Height')
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
test_mse =  mean_squared_error(ytest, reg.predict(xtest)) # testing error 
```

```python
test_mae = mean_absolute_error(ytest , reg.predict(xtest))
```
